# Citron Gelbooru Scraper Example

This notebook demonstrates how to use the citron_gelbooru_scraper library to download images and tags from Gelbooru. This app downloads the image and a txt file with the same name containing the tags for that image. 
By downloading the tags from Gelbooru there is no need to do the tagging yourself. 
Search parameters are whitespace delimited

```python
# Install the library from GitHub
!pip install git+https://github.com/citronlegacy/citron_gelbooru_scraper.git
```

```python
from citron_gelbooru_scraper.core import jelly_download

# Set your Gelbooru API credentials
api_key = "YOUR_API_KEY" #@param {type:"string"}
user_id = "YOUR_USER_ID" #@param {type:"string"}

# Define your search query
query = "catgirl rating:general -animated sort:score" #@param {type:"string"}
output_dir = "jelly_downloads" #@param {type:"string"}
max_images = 5 #@param {type:"number"}
#@markdown Auto-remove these tags from results - its best practice to remove useless tags that don't relate to your lora
auto_remove_these_tags = "tagme english_commentary" #@param {type:"string"}  

result = jelly_download(
    query=query,
    api_key=api_key,
    user_id=user_id,
    output_dir=output_dir,
    max_images=max_images,
    exclude_tags=auto_remove_these_tags
)

print(result)
```
