# Citron Gelbooru Scraper Example

This notebook demonstrates how to use the citron_gelbooru_scraper library to download images and tags from Gelbooru.

```python
from citron_gelbooru_scraper.core import jelly_download

# Set your Gelbooru API credentials
api_key = "YOUR_API_KEY" #@param {type:"string"}
user_id = "YOUR_USER_ID" #@param {type:"string"}

# Define your search query
query = "catgirl rating:general -animated sort:score" #@param {type:"string"}
output_dir = "./jelly_downloads" #@param {type:"string"}
max_images = 5

result = jelly_download(
    query=query,
    api_key=api_key,
    user_id=user_id,
    output_dir=output_dir,
    max_images=max_images
)

print(result)
```
